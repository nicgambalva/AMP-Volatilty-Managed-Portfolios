```python
# Importing the required libraries
import pandas as pd
import volman_module as vm
```

```python
df = pd.read_csv('Data/Kaggle_XAU_1d_data_2004_to_2024-09-20.csv')
print(df.head())
```

```python
df = vm.to_time_series(df, format='%Y.%m.%d')
print(df.head())
```

```python
vm.plot_time_series(df, 'Close', title='Gold Price Time Series', xlabel='Date', ylabel='Price', figsize=(10, 7), color='darkblue', save=True, save_path='gold_price.png', logscale=False, grid=False)
```

```python
vm.plot_returns(df, 'Close', title='Gold Returns Time Series', xlabel='Date', ylabel='Returns', figsize=(10, 7), color='darkred', save=True, save_path='gold_returns.png', logscale=False, grid=False)
```

```python
vm.plot_rolling_volatility(df, 'Close', title='Gold Volatility Time Series - 252 days rolling', xlabel='Date', ylabel='Volatility - 252 days rolling', figsize=(10, 7), color='darkgreen', save=True, save_path='gold_volatility.png', logscale=False, grid=False)
```

```python
vm.plot_price_volume(df, 'Close', 'Volume', title='Gold Price and Volume Time Series - Daily', xlabel='Date', ylabel_price='Close price', ylabel_volume='Volume', figsize=(10, 7), color_price='darkblue', color_volume='darkred', save=True, save_path='gold_price_volume.png', logscale=False, grid=False, ratio=0.7)
```

```python
vm.plot_returns_volatility(df, 'Close', 'Close', title='Gold Returns and Volatility Time Series - Daily', xlabel='Date', ylabel_returns='Returns', ylabel_volatility='Volatility - 252 dyas rolling', figsize=(10, 7), color_returns='darkblue', color_volatility='darkred', save=True, save_path='gold_returns_volatility_daily.png', logscale=False, grid=False, ratio=0.7, window=252)
```
